```python
light = QuantumRegister(9,'light')
push = QuantumRegister(9,'push')
out = QuantumRegister(1,'auxillary')
c = ClassicalRegister(9)

lights = [0, 1, 1, 1, 0, 0, 1, 1, 1] #input array
#push_list = [[0,1,3],[1,0,3,4],[2,1,5],[3,0,4],[4,1,3,7,5],[5,2,4,8],[6,3,7],[7,6,4,8],[8,7,5]]

qc = QuantumCircuit(push,light,out,c)

#applying input to circuit
i=0
for i in range(9):
    if lights[i] == 1:
        qc.x(light[i])
    i+=1
    
#applying hadamard to push qubits
k=0
for k in range(9):
    qc.h(push[k])
    k+=1
#push_list = [[0,1,3],[1,0,3,4],[2,1,5],[3,0,4],[4,1,3,7,5],[5,2,4,8],[6,3,7],[7,6,4,8],[8,7,5]]
#XOR operation
def XOR(qc,push,light):
    qc.cx(push[0],light[0])
    qc.cx(push[1],light[0])
    qc.cx(push[3],light[0])
    qc.cx(push[1],light[1])
    qc.cx(push[2],light[1])
    qc.cx(push[0],light[1])
    qc.cx(push[4],light[1])
    qc.cx(push[1],light[2])
    qc.cx(push[2],light[2])
    qc.cx(push[5],light[2])
    qc.cx(push[0],light[3])
    qc.cx(push[3],light[3])
    qc.cx(push[4],light[3])
    qc.cx(push[1],light[4])
    qc.cx(push[3],light[4])
    qc.cx(push[4],light[4])
    qc.cx(push[5],light[4])
    qc.cx(push[7],light[4])
    qc.cx(push[5],light[5])
    qc.cx(push[2],light[5])
    qc.cx(push[4],light[5])
    qc.cx(push[8],light[5])
    qc.cx(push[3],light[6])
    qc.cx(push[6],light[6])
    qc.cx(push[7],light[6])
    qc.cx(push[7],light[7])
    qc.cx(push[6],light[7])
    qc.cx(push[4],light[7])
    qc.cx(push[8],light[7])
    qc.cx(push[5],light[8])
    qc.cx(push[7],light[8])
    qc.cx(push[8],light[8])
    return qc
XOR(qc,push,light)
qc.x(light)
#flipping auxillary bit
qc.mct(light,out)
qc.x(light)
#uncomputing step
XOR(qc,push,light)
qc.barrier()

def diffuser(n):
    qc = QuantumCircuit(n)
    for qubit in range(n):
        qc.h(qubit)
    for qubit in range(n):
        qc.x(qubit)
    qc.h(n-1)
    qc.mct(list(range(n-1)),n-1)
    qc.h(n-1)
    for qubit in range(n):
        qc.x(qubit)
    for qubit in range(n):
        qc.h(qubit)
    return qc
qc.append(diffuser(9),push)

qc.measure(push,c)
qc.reverse_bits()
qc.draw()
```
